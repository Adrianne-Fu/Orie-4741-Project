```python
import pandas as pd
import numpy as np
```

```python
housePrice = pd.read_csv('Data/housePrice.csv')
unemployment = pd.read_csv('Data/unemployment.csv')
hospital = pd.read_csv('Data/Hospital.csv')
```

```python
Year = housePrice.Date.str.split('-', n=2, expand=True)[0]
Month = housePrice.Date.str.split('-', n=2, expand=True)[1]
```

```python
housePrice['Year'] = Year.astype(int)
housePrice['Month'] = Month.astype(int)
```

```python
housePrice = housePrice.rename(columns={'RegionName': 'County'})
housePrice['County'] = housePrice['County'].str.rsplit(' ', n=1, expand=True)[0]
```

```python
unemployment = unemployment.rename(columns={'RegionName': 'County'})
unemployment['County'] = unemployment['County'].str.rsplit(' ', n=1, expand=True)[0]
```

```python
Date = pd.to_datetime(unemployment['Date'])
Year = Date.astype(str).str.split('-', 2, expand=True)[0]
Month = Date.astype(str).str.split('-', 2, expand=True)[1]
unemployment['Year'] = Year.astype(int)
unemployment['Month'] = Month.astype(int)
```

```python
unemployment = unemployment.drop(['Date'], axis=1)
```

```python
housePrice_unemployment = pd.merge(housePrice, unemployment, how='left', on=['State','County', 'Year', 'Month'])
```

```python
hospital = hospital.rename(columns={'County Name': 'County'})
hospital.County = hospital.County.str.capitalize()
```

```python
housePrice_unemployment_hospital = pd.merge(housePrice_unemployment, hospital, how='left', on=['State', 'County'])
```
