```python
cleanData = pd.read_csv('data/cleaned_data.csv')
selectedFeeatures = cleanData[['ZHVI', 'Year', 'Month', 'UnemploymentRate', '#hospital', 'hospital_rating', 'private_schools_count', 'public_schools_count', 'CPI', 'Interest Rate',
       'inc', 'inc_ineql', 'BD', 'highschool', 'migr', 'residt']]
corrMatrix = selectedFeeatures.corr(method='spearman')

mask = np.triu(np.ones_like(corrMatrix))
heatmap = sns.heatmap(corrMatrix, square=True, mask=mask, cmap='BrBG', vmin=-1, vmax=1)
heatmap.set_title('Correlation Heatmap of Variables')
```
